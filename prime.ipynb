# 以下代码用于研究素数和伪素数的规律

1. 通过修改 prime 和 prime_nxt 来调整研究范围；
1. 不同的数字规律使用不同颜色标识；

```python
import pandas as pd

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime = [2, 3, 5, 7, 11]
prime_nxt = 13

target = 1
for i in prime:
    target *= i
    
next_calc = target    
target = max(512, (target + prime_nxt + 23))//24*24 + 2
# Generate the first 144 natural numbers starting from 2
numbers = list(range(2, target))

# Reshape the list of numbers into a DataFrame with 24 columns
df = pd.DataFrame([numbers[i:i+24] for i in range(0, len(numbers), 24)])

# Function to highlight numbers
def highlight_numbers(val):
    styles = []
    if val in prime:
        styles.append('background-color: brown')
    else:
        for i in prime:
            if val % i == 0:
                styles.append('color: gray')
                break            
        else:
            if is_prime(val):
                styles.append('background-color: lightgreen')
            else:
                styles.append('background-color: green')
    
    if (prime[-1] < val < prime[-1] + next_calc + 1):
        if val < (prime[-1] + 2) **2:
            styles.append('border: 2px dashed blue')
        else:
            styles.append('border: 2px dashed red')

    return '; '.join(styles)

pred_prime_num = 0
pred_tprime_num = 0
pred_prime_last = 0
pred_featrue = []

def check_numbers(val):
    global pred_prime_num, pred_tprime_num, pred_prime_last
    if val not in prime:
        for i in prime:
            if val % i == 0:
                break
        else:            
            if (prime[-1] + 1 < val < prime[-1] + next_calc + 2):
                if pred_prime_last + 2 == val:
                    pred_tprime_num += 1
                pred_prime_last = val
                pred_prime_num += 1
                pred_featrue.append(val)
    
for x in numbers:
    check_numbers(x)

pred_featrue = pred_featrue[:(len(pred_featrue) - 1) // 2]    
for i, val in enumerate(pred_featrue):
    if i > 0 and i % 16 == 0:
        print()
    print(val, end=' ')
print()


print('Prime predict numbers:', pred_prime_num)
print('Twin prime predict numbers:', pred_tprime_num)

# Apply the highlighting function
styled_df = df.style.map(highlight_numbers)
styled_df

```
